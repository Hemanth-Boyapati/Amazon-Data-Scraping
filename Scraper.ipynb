```python
# Function to scrape product information from a single page
def scrape_page(url):
    headers = {"User-Agent": get_user_agents(), "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}
    res = requests.get(url,headers=headers)
    soup = BeautifulSoup(res.content, 'html.parser')
    # print(soup.prettify())
    product_description=""
    try:
      product_description_div = soup.find('div', class_='a-section a-spacing-medium a-spacing-top-small')
      product_description_list =product_description_div.ul.find_all('li')
      
      for i in product_description_list:
        product_description+=i.span.text.strip()
    except:
      print('Could not retreive data')
    print(product_description)
    asin=""
    manufacturer=""
    try:
      product_details_table = soup.find('table',id='productDetails_detailBullets_sections1')
      # print(product_details_table)
      if product_details_table:
        product_details_rows =product_details_table.find_all('tr')
        if product_details_rows:
          for i in product_details_rows:
            if i.find('th').text.strip()=="ASIN":
              asin+=i.find('td').text.strip()
        product_manufacture_table = soup.find('table',id='productDetails_techSpec_section_1') 
        if product_manufacture_table: 
          product_manufacture_rows =product_manufacture_table.find_all('tr')
          if product_manufacture_rows:
            for i in product_manufacture_rows:
              if i.find('th').text.strip()=="Manufacturer":
                manufacturer+=i.find('td').text.strip()
                break
      else:
        product_details_table = soup.find('div',id='detailBullets_feature_div')
        product_details_rows=product_details_table.ul.find_all('li')
        c=0
        for i in product_details_rows:
          if i.span.find('span',class_='a-text-bold').text.find("ASIN")>=0:
            text=i.span.find_all('span')
            # print(text)
            for j in text:
                try:
                  j['class']
                except:
                  asin+=j.text.strip()                
          if i.span.find('span',class_='a-text-bold').text.find("Manufacturer")>=0 and c==0:
            c=1
            text=i.span.find_all('span')
            # print(text)
            for j in text:
                try:
                  j['class']
                except:
                  manufacturer+=j.text.strip() 
    except:
      print('Could not retreive data')
    print(asin)
    print(manufacturer)


    # manufacturer = soup.find('span', class_='a-unordered-list a-nostyle a-vertical a-spacing-none detail-bullet-list').text.strip()

    # Combine the scraped data
    product_data = {
        'Product URL': url,
        'Product Description': product_description,
        'ASIN': asin,
        'Manufacturer': manufacturer
    }

    return product_data
```
